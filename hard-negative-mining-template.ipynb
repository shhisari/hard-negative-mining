```python
D_train, lb_train = get_pos_and_random_neg(feat_extractor, "train", num_img=2500)
D_test, lb_test = get_pos_and_random_neg(feat_extractor, "validation", num_img=500)
```

```python
image_size = (256, 256)
# DATA_DIR = "ContactHands/"

# images_pool = DATA_DIR + "JPEGImages/"
# annotations_pool = DATA_DIR + "Annotations/"

def get_primal_objective(X, y, w, b, C):

    d = 1 - y * (np.dot(X, np.transpose(w) + b))
    d[d < 0] = 0
    cost = np.sum(d) * C
    objective = 1 / 2 * np.dot(w, np.transpose(w)) + cost
    return objective

# add hard negative examples
def get_B(clf):
    
    hard_negatives = []
    num = 100
    model2 = prepare_second_stream()
    images_pool = "ContactHands/JPEGImages/"
    annotations_pool = "ContactHands/Annotations/"
    dataset_file = "ContactHands/ImageSets/Main/{}.txt".format("train")
    with open(dataset_file, "r") as f:
        dataset = f.read().splitlines()
    dataset = (dataset if num == -1 else random.sample(dataset, num))
    
    for ann_name in tqdm.tqdm(dataset):
        # Get annotation and corresponding image
        xml_path = os.path.join(annotations_pool, ann_name + ".xml")
        image_file, boxes = read_content(xml_path)
        image_path = os.path.join(images_pool, image_file)
        # If image in dataset (train or test)
        if image_file.split(".jpg")[0] in dataset:
            # Read image
            img = cv2.imread(image_path)
            img = cv2.resize(img, image_size)
            
            rects = detect(img, feat_extractor, clf)
            for box in boxes:
                iou = get_iou(box, rects)
                rects = rects[iou < 0.2]
                if len(rects) == 0:
                    break

            for i in range(20):
                inference_second_stream(feat_extractor, img)
                feats = feat_extractor.extract_features([rects[i][0:4]])
                feats = feats.detach().to('cpu').numpy()
                hard_negatives.append(feats)

    return hard_negatives

def hard_negative_mining(D_train, lb_train, D_val, lb_val, C=1, max_iter=10):

    objs = []
    APs = []

    # build a svm pipeline with scaler
    scaler = StandardScaler()
    clf_i = svm.LinearSVC(max_iter=1000)
    # fit training data
    clf_i.fit(D_train, lb_train)

    # get support vectors
    decision_function = clf_i.decision_function(D_train)
    support_vectors_idx = np.where(np.abs(decision_function) <= 1 + 1e-15)[0]

    # calculate objective function
    obj_i = get_primal_objective(D_train, lb_train, clf_i.coef_, clf_i.intercept_, 1)
    # X, y, w, b, C

    pred_i = clf_i.predict(D_val)
    acc_i = metrics.accuracy_score(lb_val, pred_i)
    ap_i = metrics.average_precision_score(lb_val, pred_i)
    
    objs.append(obj_i)
    APs.append(ap_i)

    print("OBJECTIVE", obj_i)
    print("ACCURACY ", acc_i)
    print("MAP ", ap_i)

    # iterate for max_iter
    for i in range(1, max_iter):

        # We need to update training data by removing non support vectors for negative examples.
        # i.e we add only positive examples and support vectors for negative examples
        A = []
        A_lb = []
        # find all positives i.e SVs and nonSVs
        for i, j in enumerate(zip(D_train, lb_train)):
            if j[1] == 1:
               A.append(j[0])
               A_lb.append(1)

        # find SVs for negatives
        decision_function = clf_i.decision_function(D_train)
        support_vectors_idx = np.where(np.abs(decision_function) <= 1 + 1e-15)[0]
        # indices_SV = clf_i.support_
        SV_D_train = D_train[support_vectors_idx]
        SV_lb_train = lb_train[support_vectors_idx]

        for i, j in enumerate(zip(SV_D_train, SV_lb_train)):
            if j[1] == -1:
               A.append(j[0])
               A_lb.append(-1)

        # get hard negatives
        B = get_B(clf_i)
        B_lb = []   
        new_D_train = A
        for i in range(len(B) - 1):
            new_D_train = np.concatenate((new_D_train, B[i]), axis=0)
            B_lb.append(-1)

        # update new_D_train
        new_lb_train = A_lb + B_lb
        new_lb_train = np.array(new_lb_train)
        
        # retrain model
        clf_i = svm.LinearSVC(max_iter=1000)
        clf_i.fit(new_D_train, new_lb_train)

        decision_function = clf_i.decision_function(new_D_train)
        support_vectors_idx = np.where(np.abs(decision_function) <= 1 + 1e-15)[0]
        obj_i = get_primal_objective(new_D_train, new_lb_train, clf_i.coef_, clf_i.intercept_, 1)

        pred_i = clf_i.predict(D_val)
        acc_i = metrics.accuracy_score(lb_val,pred_i)
        ap_i = metrics.average_precision_score(lb_val, pred_i)

        objs.append(obj_i)
        APs.append(ap_i)

        print("OBJECTIVE", obj_i)
        print("ACCURACY ", acc_i)
        print("MAP ", ap_i)

        # update data for next iteration
        D_train = new_D_train
        lb_train = new_lb_train

    return objs, APs, clf_i
```

```python
objs, APs, clf = hard_negative_mining(D_train, lb_train, D_test, lb_test, 1, 10)
```
